## Missing values analysis

#### Significant portion of missing values

```python
na_percent = df.isna().sum()
na_percent = na_percent[na_percent > 0].sort_values(ascending=False)
na_percent
```

Detailed look of the missing values

```python
for col in na_percent.index:
    print(df[col].value_counts())
    print("Missing counts:", df[col].isna().sum())
    print("Missing percentage:", df[col].isna().sum()/df.shape[0], end='\n\n\n')
```

#### Should 'Casualty_severity' be dropped?

Null values in Casualty_severity also cause Casualty_class, Age_band_of_casualty, Sex_of_casualty to be NULL

```python
df[df['Casualty_severity'].isna()].sample(5).T
```

```python
casualty_na_df = df[df['Casualty_severity'].isna()]
casualty_na_df = casualty_na_df.isna().sum()
casualty_na_df[casualty_na_df > 0].sort_values(ascending=False)
```

Hypothesis: Might have a correlation with Accident_severity?

**H0: There is no correlation between Casualty_severity with Accident_severity**<br>
**H1: There is a correlation between Casualty_severity with Accident_severity** _(Should apply other techniques to fill null values)_

```python
for col in ['Casualty_severity', 'Accident_severity']:
    fig = plt.figure(figsize=(16, 4))
    df[col].hist(bins=50)
    plt.title(col)
    plt.show()
```

```python
crosstab_result = pd.crosstab(index=df['Casualty_severity'], columns=df['Accident_severity'])
plt.figure(figsize=(8,6))
sns.heatmap(crosstab_result, annot=True, fmt='g')
plt.show()
```

```python
from scipy.stats import chi2_contingency

ChiSqResult = chi2_contingency(crosstab_result)

print('The P-Value of the Chi-2 Test is:', ChiSqResult[1])
```

_P-value >> 0.1, cannot reject H0_ <br>
**=> Null entries in Casualty_severity should be dropped since they cause other features to be NULL**

```python
df.dropna(subset=['Casualty_severity'], axis=0, inplace=True)
```

#### Set NULL to 'Unknown'
With the columns already has the 'unknown' field, let set the null values to be 'unknown' as well

```python
for col in na_percent.index:
    if 'Unknown' in df[col].unique():
        df[col] = df[col].fillna('Unknown')
    elif 'unknown' in df[col].unique():
        df[col] = df[col].fillna('unknown')
```

#### Let's see the missing data again

```python
na_percent = df.isna().sum()/df.shape[0]
na_percent = na_percent[na_percent > 0].sort_values(ascending=False)

for col in na_percent.index:
    print(df[col].value_counts())
    print("Missing counts:", df[col].isna().sum(), end='\n\n\n')
```

__Drop__: Defect_of_vehicle, Fitness_of_casuality (high skewness) <br>
__Replace with mode__: Others

```python
df.drop(columns=['Defect_of_vehicle', 'Fitness_of_casuality'], inplace=True)
for col in ['Type_of_vehicle', 'Owner_of_vehicle', 'Road_surface_type', 'Road_allignment']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

### Transform Time

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
df.drop(columns=['Time'], inplace=True)
```
